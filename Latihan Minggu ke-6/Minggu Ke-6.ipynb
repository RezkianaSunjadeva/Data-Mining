## Missing Value

```python
data['MinTemp']=data['MinTemp'].fillna((data['MinTemp'].mean()))
data['MaxTemp']=data['MaxTemp'].fillna((data['MaxTemp'].mean()))
data['Rainfall']=data['Rainfall'].fillna((data['Rainfall'].mean()))
data['Evaporation']=data['Evaporation'].fillna((data['Evaporation'].mean()))
data['Sunshine']=data['Sunshine'].fillna((data['Sunshine'].mean()))
data['WindGustSpeed']=data['WindGustSpeed'].fillna((data['WindGustSpeed'].mean()))
data['WindSpeed9am']=data['WindSpeed9am'].fillna((data['WindSpeed9am'].mean()))
data['WindSpeed3pm']=data['WindSpeed3pm'].fillna((data['WindSpeed3pm'].mean()))
data['Humidity9am']=data['Humidity9am'].fillna((data['Humidity9am'].mean()))
data['Humidity3pm']=data['Humidity3pm'].fillna((data['Humidity3pm'].mean()))
data['Pressure9am']=data['Pressure9am'].fillna((data['Pressure9am'].mean()))
data['Pressure3pm']=data['Pressure3pm'].fillna((data['Pressure3pm'].mean()))
data['Cloud9am']=data['Cloud9am'].fillna((data['Cloud9am'].mean()))
data['Cloud3pm']=data['Cloud3pm'].fillna((data['Cloud3pm'].mean()))
data['Temp9am']=data['Temp9am'].fillna((data['Temp9am'].mean()))
data['Temp3pm']=data['Temp3pm'].fillna((data['Temp3pm'].mean()))
data['RISK_MM']=data['RISK_MM'].fillna((data['RISK_MM'].median()))
np.sum(data.isnull())
```

```python
from scipy.stats import mode
data['WindGustDir']=data['WindGustDir'].fillna(data['WindGustDir']).dropna().mode()[0]
data['WindDir9am']=data['WindDir9am'].fillna(data['WindDir9am']).dropna().mode()[0]
data['WindDir3pm']=data['WindDir3pm'].fillna(data['WindDir3pm']).dropna().mode()[0]
data['RainToday']=data['RainToday'].fillna(data['RainToday']).dropna().mode()[0]
data['RainTomorrow']=data['RainTomorrow'].fillna(data['RainTomorrow']).dropna().mode()[0]
np.sum(data.isnull())
```

## Outliers

```python
from scipy import stats
data1=data.drop(['WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],1)
z = np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
datahapusoutlier = data[(z < 3).all(axis=1)]
print(data.shape)
```

```python
datahapusoutlier.head(11)
```

## Transformasi

```python
datadrop=datahapusoutlier.drop(['WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow','Date','Location'],1)
from sklearn.preprocessing import normalize
norm = normalize(datadrop, norm = 'l2')
norm = pd.DataFrame(norm)
norm.tail()
```

```python
transformasi_sigmoidal = (1-2.718281828**(-norm))/(1+2.718281828**(-norm))
transformasi_sigmoidal = pd.DataFrame(transformasi_sigmoidal)
print(transformasi_sigmoidal)
```

```python
# Transformasi Data Kategorikal
data['RainToday'].replace({'No':0, 'Yes':1},inplace=True)
data['RainTomorrow'].replace({'No':0, 'Yes':1},inplace=True)
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
aa=le.fit(data.iloc[0:,3])
a=le.transform(data.iloc[0:,3])
bb=le.fit(data.iloc[0:,5])
b=le.transform(data.iloc[0:,5])
cc=le.fit(data.iloc[0:,6])
c=le.transform(data.iloc[0:,6])
data['WindGustDir']=a
data['WinDir9am']=b
data['winDir3pm']=c
data.head()
```
