## Data Integration

#### Data berasal dari BPS (Badan Pusat Satistik) yaitu data tentang produksi barang tambang mineral di Indonesia dari tahun 1996-2015.

```python
import pandas as pd 
import numpy as np
```

```python
Hasil1 = pd.read_csv("Produksi1.csv", sep=";")
Hasil2 = pd.read_csv("Produksi2.csv", sep=";")
Hasil3 = pd.read_csv("Produksi3.csv", sep=";")
```

```python
Hasil_completed = pd.merge(Hasil1, Hasil2, on='Tahun')
Hasil_completed2 = pd.merge(Hasil_completed, Hasil3, on='Tahun')
Hasil_completed2
```

```python
Hasil4 = pd.read_csv("Produksi4.csv", sep=";")
Hasil5 = pd.read_csv("Produksi5.csv", sep=";")

Hasil_completed3 = pd.concat([Hasil4, Hasil5])
Hasil_completed3
```
